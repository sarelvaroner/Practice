```python
DELIVERY_SIZE = 100
DELIVERIES = "deliveries"
SALES = "sales"
STORES_URL = "https://docs.google.com/spreadsheets/d/e/2PACX-1vREq5BAnlzZL1X0v3PThJI7HtzoCup6Y14X-JC1o0LERTHqYE3rYmMzOBoTHo9FvKLHNat1tjhNDEjr/pub?gid=1671734850&single=true&output=csv"
STORES = "stores"
EVENTS = "events"
ARTISTS = "artists"
SONG = "songs"
USERS = "users"
STORES_DB_FILE = f'{STORES}.db'
EVENTS_DB_FILE = f'{EVENTS}.db'
```

```python
stores_series  = pd.read_csv(STORES_URL)
```

```python
v. The store can only sell stock when it is open and is never open continuously  over 2 days. 
```

```python
a = stores_series.groupby("store")
def aa(df):
    df.index = pd.to_datetime(df["date"])
    df.sort_index(inplace=True)
    df["prev_amount"] = df["opening_stock"].shift(1).fillna(df["opening_stock"][0])
    df["change"] =  df["opening_stock"] - df["prev_amount"] 
    df["min_deliveries"] = (df[df["change"] > 0]["change"]  / DELIVERY_SIZE).apply(np.ceil) * DELIVERY_SIZE
    df["min_sales"] =  df["min_deliveries"] - df["change"]
    df.drop(inplace=True,labels=["date","prev_amount","opening_stock"], axis=1)  
    df['consecutive_deliveries'] = (df['min_deliveries'].notna() & (df['min_deliveries'].shift(1).notna() | df['min_deliveries'].shift(-1).notna())) & df["min_sales"].notna()  

    return df
a.apply(aa)
# print(a)
```

```python
SESSIONS = "sessions"
create_events_table_query = f"""CREATE TABLE IF NOT EXISTS fct_{EVENTS} (
        id INTEGER PRIMARY KEY AUTOINCREMENT,
        session_id VARCHAR(10),
        time_stemp VARCHAR(10)
        http_method VARCHAR(10),
        http_status INTEGER,
        page VARCHAR(10),
        FOREIGN KEY(session_id) REFERENCES {SESSIONS}(id)

        );"""

create_sessions_table_query = f"""CREATE TABLE IF NOT EXISTS fct_{SESSIONS} (
                                id INTEGER PRIMARY KEY AUTOINCREMENT,
                                authoritation VARCHAR(10)),
                                user_id VARCHAR(10)
                                level VARCHAR(10),
                                items_in_session INTEGER,
                                location VARCHAR(10),
                                user_agaent VARCHAR(10),
                                user_id INTEGER,
                                FOREIGN KEY(user_id) REFERENCES {USERS}(id)

                                
                                ;"""


create_users_table_query = f"""CREATE TABLE IF NOT EXISTS {USERS} (
                                id INTEGER PRIMARY KEY AUTOINCREMENT,
                                first_name VARCHAR(10),
                                last_name VARCHAR(10),
                                gender VARCHAR(10),
                                ragistation_date VARCHAR(10)
                                
                                
                                
                                ;"""




create_artists_table_query = f"""CREATE TABLE IF NOT EXISTS {ARTISTS} (
           id INTEGER PRIMARY KEY AUTOINCREMENT,
           c2 VARCHAR(10));"""


create_songs_table_query = f"""CREATE TABLE IF NOT EXISTS {SONG} (
           id INTEGER PRIMARY KEY AUTOINCREMENT,
           c2 VARCHAR(10));"""


queries = "/n".join([create_events_table_query, create_artists_table_query, create_songs_table_query, create_users_table_query])

cursor.execute(queries)



cursor.execute(f"INSERT OR IGNORE INTO {STORES}(c1, c2) VALUES(123, '456')")
cursor.execute(f"SELECT * FROM  {STORES}").fetchall()

con.commit()
```
